```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        mid_val = arr[mid]

        if mid_val == target:
            return mid
        elif mid_val < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

df = df.sort_values(by='rate', ascending=False)

# Memasukkan data DataFrame ke dalam objek Book
book_list = [Book(row['title'], row['author'], row['genre'], row['lang'], row['rate']) for _, row in df.iterrows()]

# Input dari user
target_genre = input("Masukkan genre buku yang dicari: ")
target_lang = input("Masukkan bahasa buku yang dicari: ")
target_rate = float(input("Masukkan rate buku yang dicari: "))

# Melakukan pencarian dengan binary search
result = binary_search_books(book_list, target_genre, target_lang, target_rate)

# Menampilkan hasil pencarian
if result:
    print("Buku ditemukan:")
    for book in result:
        print(f"Title: {book.title}")
        print(f"Author: {book.author}")
        print(f"Genre: {book.genre}")
        print(f"Language: {book.lang}")
        print(f"Rate: {book.rate}")
        print("\n")
else:
    print(f"Buku dengan genre {target_genre}, bahasa {target_lang}, dan rate {target_rate} ke atas tidak ditemukan.")
```
