Binary Search

```python
import psutil

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    iterations = 0  # variabel untuk menghitung iterasi

    while low <= high:
        mid = (low + high) // 2
        mid_val = arr[mid]
        iterations += 1  # Tambah iterasi setiap kali perbandingan dilakukan

        if mid_val == target:
            return mid, iterations
        elif mid_val < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, iterations  # Jika tidak ditemukan, kembalikan -1 dan jumlah iterasi

genre = df.genre.to_list()

# Mengukur penggunaan memori sebelum eksekusi
before_memory = psutil.virtual_memory().used / 1024  # Konversi byte ke kilobyte

# Mengukur waktu sebelum eksekusi
start_time = time.time()

# Penggunaan
target = input("Masukkan genre buku yang Anda cari: ")
result, iterasi = binary_search(genre, target)
print (f'iterasi={iterasi}')
# Mengukur waktu setelah eksekusi
end_time = time.time()

# Mengukur penggunaan memori setelah eksekusi
after_memory = psutil.virtual_memory().used / 1024  # Konversi byte ke kilobyte

if result != -1:
    print(f"Element {target} found at index {result}")
else:
    print(f"Element {target} not found")
    

# Cetak waktu yang dibutuhkan untuk pencarian
print(f"Time taken for Binary Search: {end_time - start_time} seconds")

# Cetak penggunaan memori selama eksekusi dalam kilobyte
memory_usage_kb = after_memory - before_memory
print(f"Memory usage: {memory_usage_kb} kilobytes")
```

```python
def create_hash_table(genres):
    hash_table = {}
    for index, genre in enumerate(genres):
        hash_key = hash(genre)
        hash_table[hash_key] = index
    return hash_table

def hash_search(hash_table, target):
    hash_key = hash(target)
    return hash_table.get(hash_key, -1)

genre = df.genre.to_list()

# Mengukur penggunaan memori sebelum eksekusi
before_memory = psutil.virtual_memory().used / 1024  # Konversi byte ke kilobyte

# Mengukur tabel hash
hash_table = create_hash_table(genre)

# Menghitung waktu sebelum eksekusi
start_time = time.time()

# User input for the target genre
target = input("Enter the genre you are looking for: ")
result = hash_search(hash_table, target)

# Mengukur waktu setelah eksekusi
end_time = time.time()

# Mengukur penggunaan memori setelah eksekusi
after_memory = psutil.virtual_memory().used / 1024  # Konversi byte ke kilobyte

if result != -1:
    print(f"Element {target} found at index {result}")
else:
    print(f"Element {target} not found")

print(f"Time taken for Hashing-based Search: {end_time - start_time} seconds")
memory_usage_kb = after_memory - before_memory
print(f"Memory usage: {memory_usage_kb} kilobytes")
```
