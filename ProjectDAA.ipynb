```python
import csv

class Book:
    def __init__(self, title, author, isbn, year_of_publication, publisher):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.year_of_publication = year_of_publication
        self.publisher = publisher

def load_books_from_csv(file_path):
    books = []
    try:
        with open(file_path, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                book = Book(row['Book-Title'], row['Book-Author'], row['ISBN'], row['Year-Of-Publication'], row['Publisher'])
                if book.title:
                    books.append(book)
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
    except Exception as e:
        print(f"An error occurred while loading the CSV file: {e}")
    return books

def create_hash_table(books):
    hash_table = {}
    for book in books:
        hashed_title = hash(book.title.lower())
        hash_table[hashed_title] = book
    return hash_table

def search_book_by_title_hashing(hash_table, target_title):
    hashed_title = hash(target_title.lower())
    if hashed_title in hash_table:
        return hash_table[hashed_title]
    else:
        return None 

if __name__ == "__main__":
    file_path = 'C://Users/wawut/OneDrive/Documents/DAA/Books.csv'
    books = load_books_from_csv(file_path)

    # Create hash table
    hash_table = create_hash_table(books)

    # input pengguna untuk judul buku
    target_title = input("Masukkan judul buku yang Anda cari: ")

    # Implementasi pencarian dengan hashing
    result_book = search_book_by_title_hashing(hash_table, target_title)
    if result_book:
        print(f"{'Judul':<30}{'Penulis':<30}{'ISBN':<15}{'Tahun':<15}{'Penerbit'}")
        print(f"{result_book.title:<30}{result_book.author:<30}{result_book.isbn:<15}{result_book.year_of_publication:<15}{result_book.publisher}")
    else:
        print(f"Buku '{target_title}' tidak ditemukan")
```

```python
import csv
from tabulate import tabulate

class Book:
    def __init__(self, title, author, date_published, publisher, genre):
        self.title = title
        self.author = author
        self.date_published = date_published
        self.publisher = publisher
        self.genre = genre

def load_books_by_genre(file_path):
    books = {}
    try:
        with open(file_path, 'r', encoding='Latin1') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                book = Book(row['title'], row['author'], row['date_published'], row['publisher'], row['genre'])
                genre_hash = hash(book.genre)
                if genre_hash not in books:
                    books[genre_hash] = []
                books[genre_hash].append(book)
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
    except Exception as e:
        print(f"An unexpected error occurred while loading the CSV file: {e}")
    return books

def recommend_books_by_genre(books, genre, num_books=5):
    genre_hash = hash(genre)
    matching_books = books.get(genre_hash, [])

    if not matching_books:
        print(f"No books found for genre '{genre}'")
        return

    matching_books.sort(key=lambda x: x.date_published, reverse=True)

    # Prepare data for tabulate
    book_data = [(book.title, book.author, book.date_published if book.date_published else 'N/A', book.publisher, book.genre) for book in matching_books[:num_books]]

    headers = ["Judul", "Penulis", "Tahun", "Penerbit", "Genre"]
    table = tabulate(book_data, headers=headers, tablefmt="pretty")
    print(table)

if __name__ == "__main__":
    file_path = 'C://Users/wawut/OneDrive/Documents/DAA/Best.csv'
    books_by_genre = load_books_by_genre(file_path)
    genre = input("Masukkan genre buku yang Anda cari: ")

    while True:
        try:
            num_books = int(input("Berapa banyak buku yang ingin Anda rekomendasikan? "))
            break
        except ValueError:
            print("Masukkan angka yang valid.")

    recommend_books_by_genre(books_by_genre, genre, num_books)
```

```python
import csv
from prettytable import PrettyTable

class Book:
    def __init__(self, title, author, date_published, publisher, genre):
        self.title = title
        self.author = author
        self.date_published = date_published
        self.publisher = publisher
        self.genre = genre

def load_books_by_genre(file_path, genre):
    books = []
    try:
        with open(file_path, 'r', encoding='Latin1') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                book = Book(row['title'], row['author'], row['date_published'], row['publisher'], row['genre'])
                if book.title and book.genre.lower() == genre.lower():
                    books.append(book)
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
    except Exception as e:
        print(f"An error occurred while loading the CSV file: {e}")
    return books

def linear_search_by_genre(books, genre):
    matching_books = [book for book in books if book.genre.lower() == genre.lower()]
    return matching_books

def recommend_books_by_genre(file_path, genre, num_books=5):
    books = load_books_by_genre(file_path, genre)

    if not books:
        print(f"No books found for genre '{genre}'")
        return

    matching_books = linear_search_by_genre(books, genre)

    if not matching_books:
        print(f"No books found for genre '{genre}'")
        return

    print(f"\nHere are {num_books} recommended books for genre '{genre}' using linear search:")

    table = PrettyTable()
    table.field_names = ['Judul', 'Penulis', 'Tahun', 'Penerbit', 'Genre']

    for book in matching_books[:num_books]:
        table.add_row([book.title, book.author, book.date_published, book.publisher, book.genre])

    print(table)

if __name__ == "__main__":
    file_path = 'C://Users/wawut/OneDrive/Documents/DAA/Best.csv'
    genre = input("Masukkan genre buku yang Anda cari: ")
    num_books = int(input("Berapa banyak buku yang ingin Anda rekomendasikan? "))
    recommend_books_by_genre(file_path, genre, num_books)
```

```python
import csv
from tabulate import tabulate
import time

class Book:
    def __init__(self, title, author, date_published, publisher, genre):
        self.title = title
        self.author = author
        self.date_published = date_published
        self.publisher = publisher
        self.genre = genre

def load_books_by_genre(file_path):
    books = {}
    try:
        with open(file_path, 'r', encoding='Latin1') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                book = Book(row['title'], row['author'], row['date_published'], row['publisher'], row['genre'])
                genre_hash = hash(book.genre)
                if genre_hash not in books:
                    books[genre_hash] = []
                books[genre_hash].append(book)
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
    except Exception as e:
        print(f"An unexpected error occurred while loading the CSV file: {e}")
    return books

def recommend_books_by_genre(books, genre, num_books=5):
    genre_hash = hash(genre)

    # Check if the genre exists in the dataset
    if genre_hash not in books:
        print(f"No books found for genre '{genre}'")
        return

    # Time Complexity: O(n log n) (sorting)
    start_time = time.time()

    matching_books = books[genre_hash]
    matching_books.sort(key=lambda x: x.date_published, reverse=True)

    # Prepare data for tabulate
    book_data = [(book.title, book.author, book.date_published if book.date_published else 'N/A', book.publisher, book.genre) for book in matching_books[:num_books]]

    headers = ["Judul", "Penulis", "Tahun", "Penerbit", "Genre"]
    table = tabulate(book_data, headers=headers, tablefmt="pretty")
    print(table)

    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"\nWaktu yang dibutuhkan: {elapsed_time:.4f} detik")

if __name__ == "__main__":
    file_path = 'C://Users/wawut/OneDrive/Documents/DAA/Best.csv'
    books_by_genre = load_books_by_genre(file_path)
    genre = input("Masukkan genre buku yang Anda cari: ")

    while True:
        try:
            num_books = int(input("Berapa banyak buku yang ingin Anda rekomendasikan? "))
            break
        except ValueError:
            print("Masukkan angka yang valid.")

    # Time Complexity: Best case, worst case, average case are theoretical measures and not directly measurable
    recommend_books_by_genre(books_by_genre, genre, num_books)
```
